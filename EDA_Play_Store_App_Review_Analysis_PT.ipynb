Since install counts vary from tens to hundreds of millions, a boxplot on a log scale provides a clear and interpretable visualization.

#### Chart - 15: Pair Plot

````python
# ----------------------------
# Clean numeric columns
# ----------------------------

# Installs
df['Installs'] = df['Installs'].astype(str)
df['Installs'] = (
    df['Installs']
    .str.replace('+', '', regex=False)
    .str.replace(',', '', regex=False)
)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Reviews
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Price
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Size (optional but recommended)
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')

# ----------------------------
# Prepare Pairplot Data
# ----------------------------
pairplot_df = df[['Rating', 'Size', 'Installs', 'Reviews', 'Price', 'Type']].copy()

# Log transform skewed variables
pairplot_df['Log_Installs'] = np.log10(pairplot_df['Installs'] + 1)
pairplot_df['Log_Reviews'] = np.log10(pairplot_df['Reviews'] + 1)

# Remove missing values
pairplot_df = pairplot_df.dropna()

# ----------------------------
# Pair Plot
# ----------------------------
sns.pairplot(pairplot_df,
    vars=['Rating', 'Size', 'Log_Installs', 'Log_Reviews', 'Price'],
    hue='Type',
    diag_kind='hist')

plt.suptitle('Pairwise Relationships: Rating, Size, Installs, Reviews, Price',
    y=1.02,
    fontsize=16)
plt.show()
````
